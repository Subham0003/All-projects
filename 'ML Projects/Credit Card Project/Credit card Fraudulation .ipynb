```python
### Splitignt the dataset

x=df.drop(['Class'],axis=1)
```

```python
y=df['Class']
```

```python
x.head()
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.23,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
Stsc=StandardScaler()
x_train_scaled=Stsc.fit_transform(X_train)
x_test_scaled=Stsc.fit_transform(X_test)
```

```python
x_train_scaled
```

```python
# Importing LogisticRegression from scikit-learn
from sklearn.linear_model import LogisticRegression

# Creating an instance of LogisticRegression
logit_model = LogisticRegression(max_iter=100, random_state=42)

# Fitting the model to the training data
logit_model.fit(x_train_scaled, Y_train)
```

```python
# Predicting the model on the training data
y_pred_train_logit = logit_model.predict(x_train_scaled)
y_pred_test_logit = logit_model.predict(x_test_scaled)
```

```python
# Importing necessary metrics from scikit-learn
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, precision_score, recall_score, f1_score

# Displaying the Accuracy score for train and test data
print("Training Accuracy Score :", accuracy_score(Y_train, y_pred_train_logit))
print("*"*100)
print("Test Accuracy Score :", accuracy_score(Y_test, y_pred_test_logit))
```

```python
# Displaying the Classification Report
print("Training Classification Report:\n", classification_report(Y_train, y_pred_train_logit, target_names = ['Non-Fraudulent', 'Fraudulent']))
print("*"*100)
print("Test Classification Report:\n", classification_report(Y_test, y_pred_test_logit, target_names = ['Non-Fraudulent', 'Fraudulent']))
```

```python
###Lets balace the dataset
```

```python
from imblearn.under_sampling import  RandomUnderSampler
```

```python
resampled_df = RandomUnderSampler(sampling_strategy='auto', random_state=42)
x_resampled, y_resampled = resampled_df.fit_resample(x,y)
```

```python
print("Actual Dataset Count: ", y.value_counts())
print("*"*100)
print("After UnderSampling Method: ", y_resampled.value_counts())
print("*"*100)
```

```python
y_resampled.value_counts()
```

```python
X_train_resample,X_test_resample,Y_train_resample,Y_test_resample=train_test_split(x_resampled,y_resampled,test_size=0.23,random_state=42)
```

```python
# Importing LogisticRegression from scikit-learn
from sklearn.linear_model import LogisticRegression

# Creating an instance of LogisticRegression
logit_model = LogisticRegression(max_iter=100, random_state=42)

# Fitting the model to the training data
logit_model.fit(X_train_resample, Y_train_resample)
```

```python
# Predicting the model on the training data
y_pred_train_logit = logit_model.predict(X_train_resample)
y_pred_test_logit = logit_model.predict(X_test_resample)

# Importing necessary metrics from scikit-learn
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, precision_score, recall_score, f1_score

# Displaying the Accuracy score for train and test data
print("Training Accuracy Score :", accuracy_score(Y_train_resample, y_pred_train_logit))
print("*"*100)
print("Test Accuracy Score :", accuracy_score(Y_test_resample, y_pred_test_logit))

# Displaying the Classification Report
print("Training Classification Report:\n", classification_report(Y_train_resample, y_pred_train_logit, target_names = ['Non-Fraudulent', 'Fraudulent']))
print("*"*100)
print("Test Classification Report:\n", classification_report(Y_test_resample, y_pred_test_logit, target_names = ['Non-Fraudulent', 'Fraudulent']))
```

```python
cm=confusion_matrix(Y_test_resample, y_pred_test_logit)
```

```python
Labels=['Non-Fraud','Fraud']
sns.heatmap(cm,annot=True, cbar = False, cmap = 'cividis', fmt='d',xticklabels=Labels, yticklabels=Labels)
plt.title("confussion amtrix")
```

```python
### Cross valdiation with K fold method
from sklearn.model_selection import cross_val_score
```

```python
logit_model = LogisticRegression()
```

```python
training_accuracy = cross_val_score(logit_model, X_train_resample, Y_train_resample, cv=10, scoring='accuracy')

# Testing accuracy using cross-validation
testing_accuracy = cross_val_score(logit_model, X_test_resample, Y_test_resample, cv=10, scoring='accuracy')

# Print the results
print(f"Training Accuracy: {training_accuracy.mean()}")
print(f"Testing Accuracy: {testing_accuracy.mean()}")
```

```python
# Importing LogisticRegression from scikit-learn
from sklearn.ensemble import RandomForestClassifier

# Creating an instance of LogisticRegression
RFC_model = RandomForestClassifier(n_estimators=101, random_state=42)

# Fitting the model to the training data
RFC_model.fit(X_train_resample, Y_train_resample)
```

```python
# Predicting the model on the training data
y_pred_train_RFC = RFC_model.predict(X_train_resample)
y_pred_test_RFC = RFC_model.predict(X_test_resample)

# Importing necessary metrics from scikit-learn
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, precision_score, recall_score, f1_score

# Displaying the Accuracy score for train and test data
print("Training Accuracy Score :", accuracy_score(Y_train_resample, y_pred_train_RFC))
print("*"*100)
print("Test Accuracy Score :", accuracy_score(Y_test_resample, y_pred_test_RFC))

# Displaying the Classification Report
print("Training Classification Report:\n", classification_report(Y_train_resample, y_pred_train_logit, target_names = ['Non-Fraudulent', 'Fraudulent']))
print("*"*100)
print("Test Classification Report:\n", classification_report(Y_test_resample, y_pred_test_logit, target_names = ['Non-Fraudulent', 'Fraudulent']))
```

```python
from sklearn.metrics import roc_curve, auc
```

```python
fpr, tpr, thresholds = roc_curve(Y_test_resample,y_pred_test_RFC)
roc_auc = auc(fpr, tpr)

# Plot ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', label=f'Model (AUC = {roc_auc:.2f})')

# Plot the diagonal line for reference
plt.plot([0, 1], [0, 1], linestyle='--', color='gray', label='Random')

# Customize the plot
plt.title('AUC-ROC Curve for Random forest')
plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.legend()
plt.show()
```
